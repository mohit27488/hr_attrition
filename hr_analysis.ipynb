```python
#Taking backup of the original dataframe
hr_data_backup=hr_data.copy()

#Removing the columns which have a constant value and does not add anything.
hr_data=hr_data.drop(['EmployeeCount','StandardHours','EmployeeNumber'],axis=1)
```

```python
hr_data.head(1)
```

```python
# Dividing the age values into bins 
bins=pd.IntervalIndex.from_tuples([(17, 25), (25, 35), (35, 45),(45,55),(55,60)])
age_bin=pd.cut(hr_data['Age'],bins)
```

```python
hr_data['Attrition_value']=hr_data['Attrition'].map({'Yes':1,'No':0})
```

```python
hr_data=hr_data.drop(['Attrition'],axis=1)
```

```python
hr_data['Age_bin']=age_bin
```

```python
hr_age=hr_data.groupby(['Age_bin','Attrition_value']).count()
hr_age=hr_age.reset_index()
hr_age
```

```python
sns.barplot(hr_age['Age_bin'],hr_age['Age'],hue=hr_age['Attrition_value'])
```

```python
hr_data['Age_bin'].unique()
```

```python
fraction=pd.DataFrame(columns=['Age_bin','fraction'])
for n in hr_data['Age_bin'].unique():
    print(n)
    len_1=len(hr_data[(hr_data['Age_bin']==n) & (hr_data['Attrition_value']==1)])
    print(len_1)
    len_tot=len(hr_data[(hr_data['Age_bin']==n)])
    print(len_tot)
    fract=(len_1/len_tot)*100
    print(fract)
    fraction=fraction.append({'Age_bin':n,'fraction':fract},ignore_index=True)
```

```python
fraction
```

```python
sns.barplot('Age_bin','fraction',data=fraction)
```

Maximum Attrition percentage is in the age group of 17-25

```python
def fraction_attr(column):
    frc_wise=pd.DataFrame(columns=[column,'fraction'])
    for n in hr_data[column].unique():
        print(n)
        len_1=len(hr_data[(hr_data[column]==n) & (hr_data['Attrition_value']==1)])
        print(len_1)
        len_tot=len(hr_data[(hr_data[column]==n)])
        print(len_tot)
        fract=(len_1/len_tot)*100
        print(fract)
        frc_wise=frc_wise.append({column:n,'fraction':fract},ignore_index=True)
    return(frc_wise)
```

```python
a=fraction_attr('Gender')
```

Male have higher Attrition Rate as compared to Females.
